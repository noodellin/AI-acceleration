```python
import bs4
import requests
import os
```

```python
url = "http://www.grandtech.info/"
html = requests.get(url)
html
```

```python
dateDir = "outimg"
if os.path.exists(dateDir) == False:
  os.mkdir(dateDir)
```

```python
objsoup = bs4.BeautifulSoup(html.text,"lxml")
```

```python
imgtag = objsoup.select("img")
print(len(imgtag))
print(imgtag)
```

```python
for i in imgtag:
  imgurl = i.get("src")
  finurl = url + imgurl
  picget = requests.get(finurl)
  output = open(os.path.join(dateDir,os.path.basename(imgurl)), 'wb')
  for disk in picget:
    output.write(disk)
  output.close()
```

```python
import bs4
import requests
import os
from skimage import io
```

```python
Dir = "outimg"
if os.path.exists(Dir) == False:
    os.mkdir(Dir)
```

```python
url = "https://www.cartoonmad.com/"
rqt = requests.get(url)
rqt
```

```python
bsrqt = bs4.BeautifulSoup(rqt.text,"lxml")
img = bsrqt.find_all("img")
for i in range(len(img)-10,len(img)):
    imgurl = img[i].get("src")
    finurl = url + imgurl
    print(finurl)
    img_read = io.imread(finurl)
    io.imshow(img_read)
    io.show()
```

```python
def grabimg(url):                              #input url you want to grab image
    rqt = requests.get(url)                    #use requests to get url
    bsrqt = bs4.BeautifulSoup(rqt.text,"lxml") #use bs4 to get url info
    img = bsrqt.find_all("img")                #find image key word
    for i in range(int(0.5*len(img)),len(img)): #select half number image of web
        imgurl = img[i].get("src")             #get image url
        finurl = url + imgurl                  #merge web url and image url
        print(finurl)                          #show complete image url
        img_read = io.imread(finurl)           #use io to read image
        io.imshow(img_read)                    #input image you want to show
        io.show()                              #show image
```

```python
grabimg("https://www.cartoonmad.com/")
```
